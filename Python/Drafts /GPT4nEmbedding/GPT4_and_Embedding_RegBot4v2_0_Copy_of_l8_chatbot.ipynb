```python
# Define a function to count number of tokens in a collection of messages
max_response_tokens = 250
token_limit= 8192
#conversation=[]
#conversation.append(context)

def num_tokens_from_messages(messages, model="gpt-4"):
    encoding = tiktoken.encoding_for_model(model)
    num_tokens = 0
    for message in messages:
        num_tokens += 4  # every message follows <im_start>{role/name}\n{content}<im_end>\n
        for key, value in message.items():
            num_tokens += len(encoding.encode(value))
            if key == "name":  # if there's a name, the role is omitted
                num_tokens += -1  # role is always required and always 1 token
    num_tokens += 2  # every reply is primed with <im_start>assistant
    return num_tokens
```

#Semantic text search using embeddings
We can search through all regulation text semantically in a very efficient manner and at very low cost, by embedding our search query, and then finding the most similar reviews.

```python
import pandas as pd
import numpy as np

datafile_path = "cal_regs_with_embeddings.csv"

df = pd.read_csv(datafile_path)
df["embedding"] = df.embedding.apply(eval).apply(np.array)
```

```python
df
```

```python
from openai.embeddings_utils import get_embedding, cosine_similarity

# Define a search function to get the embedding of the prompt
def search_calregs(df, prompt, n, pprint=False):
    prompt_embedding = get_embedding(
        prompt,
        engine="text-embedding-ada-002"
    )
    # Add to the dataframe a column named similarity to store cosine similarity results
    df["similarity"] = df.embedding.apply(lambda x: cosine_similarity(x, prompt_embedding))
    # Get the top 3 values
    results = (
        df.sort_values("similarity", ascending=False)
        .head(n)
        .combined.str.replace(">>", ":")
        .str.replace("ยง", "Section")
    )
    if pprint:
        for r in results:
            print(r[:20000])
            print()
    results_str = "\n".join(results)
    return results_str
```

```python
# Test the search function, it works as intended
results_str = search_calregs(df, "What should a public water system do if one or more samples in the repeat sample set is total coliform-positive?", 1,  pprint=True)
```

```python
# This function will extract the prompt from user input, search the regulations for relevant text and append all results into the context list
def collect_messages(_):
    #extract the prompt from user input
    prompt =  "What should a public water system do if one or more samples in the repeat sample set is total coliform-positive?"
    #prompt = inp.value_input
    inp.value = ''
    #append the prompt to the context list
    context.append({'role':'user', 'content':f"{prompt}"})
    #search the regulations for relevant text
    results_str = search_calregs(df, prompt, 1)
    #append the results to the context list
    context.append({'role':'assistant', 'content':f"{results_str}"})
    #compute the number of tokens in the chat history
    conv_history_tokens = num_tokens_from_messages(context)
    #if history tokens are more than token limit then delete the oldest prompt
    while (conv_history_tokens+max_response_tokens >= token_limit):
        del context[1]
        conv_history_tokens = num_tokens_from_messages(context)
    #call the get completion function and pass on the ongoing context list to get a response
    response = get_completion_from_messages(context)
    #append the response to the context list
    context.append({'role':'assistant', 'content':f"{response}"})
    #show prompt and response on the panel
    panels.append(
        pn.Row('User:', pn.pane.Markdown(prompt, width=600)))
    panels.append(
        pn.Row('RegBot:', pn.pane.Markdown(response, width=600, style={'background-color': '#F6F6F6'})))

    return pn.Column(*panels)
```
